Plotando gr√°ficos simples:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(7)
y = np.random.randint(low = 1, high = 1500, size = 10) #valores inteiros pseudo-aleat√≥rios entre 1 e 1500, com 10 amostras
plt.plot(y) #plota os valores
```

```python
plt.plot(y, color = '#749187', marker = 'o', ms = 5, mec = 'k', markerfacecolor = 'w', ls = '-.') #insere primeira linha
plt.plot(y*2, marker = '+', color = 'k', ms = 5) #insere segunda linha

#R√≥tulos
plt.xlabel('Eixo X', color = 'red', size = 12)
plt.ylabel('Eixo Y')
plt.title('T√≠tulo', loc = 'left')

#linhas de grade
plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 1, alpha = 0.8)
plt.show()
```

Plotando gr√°ficos e ajustando par√¢metros:

```python
# Gerando numeros com o numpy
Dados_x = np.linspace(0, 10, 200) # retorna valores uniformemente espa√ßados dentro do determinado definido.
Dados_y = Dados_x * Dados_x

# Ajustando a figura
plt.figure(
          figsize=(10,5), # Ajustando o tamanho do gr√°fico
          facecolor='gray' # Ajustando cor de fundo
)

# Adicionando eixos
Ax = plt.axes()
Ax.set_facecolor('white') # Definindo cor de fundo no gr√°fico

#Criando o gr√°fico e passando par√¢metros para ajustes
plt.plot(
         Dados_x, Dados_y, # Dados do gr√°ficos
         color='magenta', # Cor da linha
         linewidth=2, # Espessura da linha
         markersize=5, # Tamanho do marcador
         marker='o', # Tipo do marcados
         linestyle='dashed', # Tipo da linha
         markeredgecolor='black', # Cor linha do marcador
)

# Definindo label do eixo x
plt.xlabel(
            'Eixo x', # nome do label
            fontweight='bold', # estilo
            fontsize='large', # Tamanho da fonte
            fontfamily='fantasy', # Tipo da fonte
            color='blue', # Cor da Fonte
)

# Definindo label do eixo y
# Mesmo par√¢metros pode ser aplicado do 'xlabel'
plt.ylabel( 'Eixo y' )

# Definindo um titulo
plt.title(
          'Gr√°fico Simples', # T√≠tulo do Gr√°fico
          fontweight='bold', # Estilo da fonte
          fontsize='xx-large', # Tamanho da fonte
          fontfamily='fantasy', # Tipo da fonte
          color='blue', # Cor da Fonte
)

# Definindo as legendas
plt.legend(
           'N√∫meros Aleat√≥rios', # Nome da legenda
           loc='upper center', # Posi√ß√£o da legenda
           bbox_to_anchor=(1, 0., 0.0, 0.2), # Posi√ß√£o legenda na Figura
           fontsize='large', # Tamanho do texto
           shadow=True, # Sombra na legenda
           borderpad=1, # Bordar da legenda
);
```

Subplots:

```python
np.random.seed(6)
x = np.arange(1, 11) # retorna valores uniformemente espa√ßados dentro do determinado definido.
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

plt.figure(figsize = (15, 5))
plt.suptitle('Figura', fontsize = 15)

# Grafico na 1¬∫ posi√ß√£o da Grade
plt.subplot(1, 3, 1)
plt.plot(x, y1, color = 'k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Grafico na 2¬∫ posi√ß√£o da Grade
plt.subplot(1, 3, 2)
plt.plot(x, y2, color = 'm')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# Grafico na 3¬∫ posi√ß√£o da Grade
plt.subplot(1, 3, 3)
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad = 4) # ajusta o espa√ßamento entre os subplots
plt.show()
```

```python
fig ,ax = plt.subplots(1, 3, figsize = (15, 5))

fig.suptitle('Figura')
ax[0].plot(x, y1, color = 'black')
ax[1].plot(x, y2, color = 'red')
ax[2].plot(x, y3, color = 'green')

for i in range(3):
  ax[i].set(title = f'Subplot {i + 1}', xlabel = 'Eixo X', ylabel = 'Eixo Y')
```

```python
# Definindo a dimens√£o da Grade
Linhas = 2
Colunas = 2

# Grafico na 1¬∫ posi√ß√£o da Grade
plt.subplot( Linhas, Colunas, 1 )
plt.plot( Dados_x, 'r--' )
plt.title('Grafico 1')

# Grafico 2¬∫ posi√ß√£o da Grade
plt.subplot( Linhas, Colunas, 2 )
plt.plot( Dados_y, 'g-' )
plt.title('Grafico 2')

# Grafico 3¬∫ posi√ß√£o da Grade
plt.subplot( Linhas, Colunas, 3 )
plt.plot( Dados_y , 'y-' )
plt.title('Grafico 3')

# Grafico 4¬∫ posi√ß√£o da Grade - Incluindo 2 eixos
plt.subplot( Linhas, Colunas, 4 )
plt.plot( Dados_x , 'b--' )
plt.plot( Dados_y , 'p--' )
plt.title('Grafico 4')

plt.tight_layout(pad = 1)
plt.show()
```

Gr√°fico com mini gr√°fico acoplado:


```python
# Criando uma nova firua
Figura = plt.figure()

# Definindo o tamanho do grafico
Eixo_1 = Figura.add_axes([0.1, 0.1, 0.8, 0.8]) #[left, bottom, width, height]

# Definindo o tamanho do grafico acoplado
Eixo_2 = Figura.add_axes([0.2, 0.5, 0.3, 0.3])

# Plotando os graficos
Eixo_1.plot( Dados_x, Dados_y, 'r--' )
Eixo_2.plot( Dados_y, Dados_x, 'y' );
```

Gr√°fico de dispes√£o ou Scatter plot:

(Representa√ß√µes gr√°ficas do relacionamento entre duas vari√°veis num√©ricas).

```python
x = np.random.normal(10, 0.5, 100)
y = np.random.uniform(0, 20, 100)
fig = plt.figure()
ax = plt.axes()
ax.scatter(x, y, marker = 'o', color = 'red', label = 'data1', alpha = 0.5)
ax.scatter(x*0.5, y*0.5, marker = 'v', color = 'black', label = 'data2', alpha = 0.7)
ax.legend()
```

Gr√°fico de Barras:

```python
# Criando a figura
Figura = plt.figure()

Dados_z = np.linspace(1, 200, 10)

# Definindo o Tamanho
Grafico = Figura.add_axes([0, 0, 0.7, 0.7])

# Definindo os rotulos
Rotulos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Gerando o gr√°fico
Grafico.bar(Rotulos,Dados_z);
```

///////////////////////// **Exemplo com Dados Econ√¥micos:** ////////////////////

```python
import plotly.express as px

px.data.gapminder()
```

```python
df = px.data.gapminder().query('country == "Brazil"').set_index('year')
df.head()
```

```python
plt.plot(df.index, df['gdpPercap'])
plt.title('PIB per capita do Brasil')
plt.ylabel('PIB per capita')
plt.xlabel('Tempo')
```

```python
title = 'Rela√ß√£o entre expectativa de vida e renda per capita no Brasil'
plt.figure(figsize = (12, 4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap = 'viridis')
plt.xlabel('Expectativa de vida')
plt.ylabel('Renda per capita')
plt.title(title, loc = 'left')
plt.show()
```

```python
plt.bar(x = df.index, height = df['pop'], color = 'red')
plt.title('Popula√ß√£o brasileira')
```

```python
def filtrar_continente(continente):
  df = px.data.gapminder()
  df = df[df['continent'] == continente]
  return df
```

```python
def filtrar_pais (pais, variavel):
  'Filtrar algum pa√≠s das amostras para as variav√©is "pop", "gdpPercap" e "lifeExp"'
  df = px.data.gapminder()
  df = df[df['country'] == pais][variavel]
  return df
```

```python
americas = filtrar_continente('Americas')
paises = americas['country'].unique()
plt.figure(figsize = (12, 8))

for pais in paises:
  plt.scatter( filtrar_pais (pais = pais, variavel = 'lifeExp'),
              filtrar_pais(pais = pais, variavel = 'gdpPercap'))
  plt.legend(labels = paises, loc = 'best')
  plt.title('Rela√ß√£o entre renda per capita e expectativa de vida', loc = 'left')
  plt.xlabel('Expectativa de vida')
  plt.ylabel('Renda per capita')
```
